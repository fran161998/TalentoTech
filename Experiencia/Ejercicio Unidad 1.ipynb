```python
import pandas as pd

url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
titanic = pd.read_csv(url)
```

Calcular las medidas de tendencia central para las
variables survived, pClass y Sex.


```python
#Supervivencia
mediaEdad = titanic['Survived'].mean()
medianaEdad = titanic['Survived'].median()
modaEdad = titanic['Survived'].mode()
print("Medidas de tendencia central para la variable Superivencia:")
print("Media:", mediaEdad)
print("Mediana:", medianaEdad)
print("Moda:", modaEdad)
```

```python
#pClas
mediaPclass = titanic['Pclass'].mean()
medianaPclass = titanic['Pclass'].median()
modaPclass = titanic['Pclass'].mode()
print("Medidas de tendencia central para la variable Pclass:")
print("Media:", mediaPclass)
print("Mediana:", medianaPclass)
print("Moda:", modaPclass)
```

Calcular los cuantiles de las variables edad y costo de
tiquete.

```python
print("\nCuantiles - Edad:")
print(titanic["Age"].quantile([0.25, 0.5, 0.75]))

print("\nCuantiles - Costo del tiquete (Fare):")
print(titanic["Fare"].quantile([0.25, 0.5, 0.75]))
```
