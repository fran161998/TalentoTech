```python
import pandas as pd

url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
titanic = pd.read_csv(url)
```

Visualizar las columnas que tiene el archivo y, con ayuda de
internet determinar qué significa cada una de ellas.


```python
print("Columnas disponibles:\n", titanic.columns)
titanic.head()
```

PassengerId: identificador del pasajero

Survived: 0 = no sobrevivió, 1 = sí sobrevivió

Pclass: clase del boleto (1, 2, 3)

Name: nombre

Sex: sexo

Age: edad

SibSp: número de hermanos/esposos a bordo

Parch: número de padres/hijos a bordo

Ticket: número de ticket

Fare: tarifa pagada

Cabin: cabina

Embarked: puerto de embarque (C, Q, S).

**Tipos de variables:**
Cualitativas: Sex, Embarked, Name, Ticket, Cabin
Cuantitativas: Age, Fare, SibSp, Parch
Categoricas: Pclass, Survived



Calcular el promedio de supervivencia general

```python
promedioSupervivencia = titanic["Survived"].mean()
print("El promedio de supervivencia general es:", promedioSupervivencia)
```

Calcular el promedio de supervivencia de los hombres


```python
supervivenciaHombres = titanic[titanic["Sex"] == "male"]["Survived"].mean()
print("El promedio de supervivencia de los hombres es:", supervivenciaHombres)
```

Calcular el promedio de supervivencia de las mujeres


```python
supervivenciaMujeres = titanic[titanic["Sex"] == "female"]["Survived"].mean()
print("El promedio de supervivencia de las mujeres es:", supervivenciaMujeres)
```

Según la cantidad de clases de tiquete, calcular el promedio de
supervivencia para cada uno de ellos.

```python
supervivenciaClase = titanic.groupby("Pclass")["Survived"].mean()
print("El promedio de supervivencia por clase de boleto es:\n", supervivenciaClase)
```
