```python
import pandas as pd
import matplotlib.pyplot as plt


url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
titanic = pd.read_csv(url)
```

Crear una distribución de frecuencia con las edades de los
pasajeros

```python
#Eliminas valores nulos de la columna edades
edades = titanic['Age'].dropna()

frecuenvia_edades = edades.value_counts().sort_index()

print("Distribucion frecuencias edad :",frecuenvia_edades)
```

Crear una distribución de frecuencia con los precios,
como es una variable cuantitativa,
Realizar las distribuciones por intervalos, teniendo en
cuenta 10 divisiones entre el dato


```python
precios = titanic['Fare'].dropna()
frecuenciaPrecios = precios.value_counts(bins=10).sort_index()
print("Distribucion frecuencias precios :",frecuenciaPrecios)
```

 Graficar las distribuciones empleando histogramas

```python
#histograma Edades
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.hist(edades, bins=10, color="skyblue", edgecolor="black")
plt.title("Distribución de Edades")
plt.xlabel("Intervalos de Edad")
plt.ylabel("Frecuencia")
```

```python
#histograma precios
plt.subplot(1,2,2)
plt.hist(precios, bins=10, color="skyblue", edgecolor="black")
plt.title("Distribución de Precios")
plt.xlabel("Intervalos de Precios")
plt.ylabel("Frecuencia")
```
