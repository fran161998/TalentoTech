```python
import itertools
import math
```

Espacio muestreal del experimento estadistico considerado al lanzar 2 dados


```python
dado = [1,2,3,4,5,6]
esspacioMuestral = list(itertools.product(dado,dado))
print(esspacioMuestral)
```

qu probabilidad tiene cada evento en el espacio muestral de ocurrir

```python
probabilidadEvento = 1/len(esspacioMuestral)
print(probabilidadEvento)
```

Cual es la probabilidad que la suma de los dos dados sea 8

```python
casoFavorable = 0
for evento in esspacioMuestral:
  if sum(evento) == 8:
    casoFavorable += 1
probabilidad = casoFavorable/len(esspacioMuestral)
print(probabilidad)
```

si se tiene un club de 50 miembros y se quiere dar un premio a 4 de ellos Cuantas posibilidades hay que cuatro de ellos ganen el premio?

```python
miembros = 50
ganadores = 4
combinaciones = math.comb(miembros,ganadores)
print(combinaciones)
probabilidadGanar = 1/combinaciones
print("La probabilidad de 4 de ellos ganen el premio es de",probabilidadGanar)
```
