```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm

#el dataset de iris ya viene incluido en seaborn entonces solamente importamos
iris = sns.load_dataset("iris")
print("Primeras filas del dataset Iris:\n", iris.head())
```

Separamos los 3 tipos de iris (setosa, virsicolor y virgiinica)

```python
setosa = iris[iris["species"] == "setosa"]
versicolor = iris[iris["species"] == "versicolor"]
virginica = iris[iris["species"] == "virginica"]
```

Hsitograma de longitud del sepalo y curva gausiana


```python
especies = {"Setosa": setosa, "Versicolor": versicolor, "Virginica": virginica}

for nombre, df in especies.items():
    data = df["sepal_length"]

    media = data.mean()
    desviacion = data.std()

    plt.figure(figsize=(7,4))
    # Histograma
    plt.hist(data, bins=10, color="skyblue", edgecolor="black", density=True, alpha=0.7)
    plt.title(f"{nombre} - Longitud del Sépalo\nMedia={media:.2f}, Desv.Std={desviacion:.2f}")
    plt.xlabel("Longitud del sépalo (cm)")
    plt.ylabel("Densidad")
    # Curva gausiana
    x = np.linspace(data.min(), data.max(), 100)
    pdf = norm.pdf(x, media, desviacion)
    plt.plot(x, pdf, "r-", linewidth=2, label="Curva Gaussiana")
    plt.legend()
    plt.show()
```

Repeticion del punto 3 y 4 con los datos del ancho del sepalo

```python
def graficarVariable(df, columna, nombre_especie):
    data = df[columna]
    media = data.mean()
    desviacion = data.std()

    plt.figure(figsize=(7,4))
    plt.hist(data, bins=10, color="lightgreen", edgecolor="black", density=True, alpha=0.7)
    plt.title(f"{nombre_especie} - {columna}\nMedia={media:.2f}, Desv.Std={desviacion:.2f}")
    plt.xlabel(columna)
    plt.ylabel("Densidad")

    x = np.linspace(data.min(), data.max(), 100)
    pdf = norm.pdf(x, media, desviacion)
    plt.plot(x, pdf, "r-", linewidth=2, label="Curva Gaussiana")
    plt.legend()
    plt.show()
```

```python
for nombre, df in especies.items():
    graficarVariable(df, "sepal_width", nombre)
```

Repeticion del punto 3 y 4 con la longitud del petalo y el ancho del petalo

```python
for nombre, df in especies.items():
    graficarVariable(df, "petal_length", nombre)
    graficarVariable(df, "petal_width", nombre)
```
