```python

import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
titanic = pd.read_csv(url)
```

```python
titanic.columns
```

limpiamos los valores nulos de edad y supervivientes

```python
titanic = titanic.dropna(subset=['Age', 'Survived'])
```

Estadisticas generales de las edades

```python
print("Desviación estándar:", titanic["Age"].std())
print("Varianza:", titanic["Age"].var())
print("Rango intercuartil (IQR):", titanic["Age"].quantile(0.75) - titanic["Age"].quantile(0.25))
```

Estadisticasd de las edades de ls supervivientes

```python
supervivientesEdades = titanic.groupby("Survived")["Age"]
for esrado, edades in supervivientesEdades:
    print(f"Estadísticas para la persona {esrado}:")
    print("\nSuperviviente" if esrado == 1 else "\nNo superviviente")
    print("Desviación estándar:", edades.std())
    print("Varianza:", edades.var())
    print("Rango intercuartil", edades.quantile(0.75) - edades.quantile(0.25))
```

Graficas de cajas y bigotes

```python
plt.figure(figsize=(8,6))
sns.boxplot(data=titanic, x="Survived", y="Age")
plt.xticks([0,1], ["No sobrevivio", "Sobrevivio"])
plt.title("Distribución de edades según supervivencia")
plt.show()
```
