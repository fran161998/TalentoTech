```python
import pandas as pd
import numpy as np


alturas = [
    229, 234, 218, 216, 219, 211, 223, 238, 233, 219,
    227, 224, 226, 222, 209, 222, 219, 221, 230, 224,
    230, 208, 231, 218, 229, 231, 222, 214, 215, 226,
    211, 226, 223, 226, 217, 221, 220, 230, 213, 233
]


datos = pd.Series(alturas)


tabla = datos.value_counts().sort_index()
fi = tabla.values
fr = fi / fi.sum()
Fi = np.cumsum(fi)
Fr = np.cumsum(fr)

tabla_frecuencias = pd.DataFrame({
    "Altura": tabla.index,
    "fi": fi,
    "fr": fr.round(3),
    "Fi": Fi,
    "Fr": Fr.round(3)
})

print("Tabla de frecuencias:")
display(tabla_frecuencias)
```

A partir de la tabla, identificar la moda, el mínimo y el
máximo. Calcular los tres cuartiles con la
distribución de frecuencias de las alturas de las
plantas

```python
minimo = tabla_frecuencias["Altura"].min()
maximo = tabla_frecuencias["Altura"].max()
moda = tabla_frecuencias.loc[tabla_frecuencias["fi"] == tabla_frecuencias["fi"].max(), "Altura"].values[0]
print(f"Moda: {moda}")
print(f"Mínimo: {minimo}")
print(f"Máximo: {maximo}")

#Cuartiles
q1 = datos.quantile(0.25)
q2 = datos.quantile(0.5)
q3 = datos.quantile(0.75)
print(f"Q1: {q1}")
print(f"Q2: {q2}")
print(f"Q3: {q3}")
```

Calcular la media aritmética de las alturas de las
plantas.

```python
media = np.mean(datos)
print(f"Media: {media}")
```

Se tienen los datos de una segunda plantación, para
la cual las flores se separan más (baja densidad por
área) en el suelo. Para estos datos, realizar los
cálculos de los numerales 1, 2 y 3 y comparar los
resultados con los datos de las 40 plantas iniciales.
Teniendo en cuenta que los 40 datos iniciales son
de girasole sembrados en un terreno con alta
densidad por área, concluir cual proceso es mejor
basados en la evidencia.


```python
segundaPlantacion = [193,197,198,199,201,202,204,205,207,208,209,210,212,213,214,215,216,217,219,220,221,222,224,225,226]
```

```python
def analizar(datos, nombre):
    serie = pd.Series(datos)
    resultados = {
        "Mínimo": serie.min(),
        "Máximo": serie.max(),
        "Moda": serie.mode().tolist(),
        "Media": serie.mean(),
        "Q1": serie.quantile(0.25),
        "Mediana (Q2)": serie.median(),
        "Q3": serie.quantile(0.75)
    }
    print(f"Resultados para {nombre}:")
    for k,v in resultados.items():
        print(f"{k}: {v}")
    return resultados

primeraPlantacion = analizar(datos, "Primera Plantacion")
segundaPlantacion = analizar(segundaPlantacion, "Segunda Plantacion")
```

```python
print("\nComparación entre plantaciones:")
print(f"Densidad primera plantacion: {primeraPlantacion['Media']:.2f}")
print(f"Densidad segnda plantacion: {segundaPlantacion['Media']:.2f}")

if segundaPlantacion['Media'] > primeraPlantacion['Media']:
    print("\n la segunda plantacion produce girasoles mas altos")
else:
    print("\n la primera plantacion produce girasoles mas altos")
```
