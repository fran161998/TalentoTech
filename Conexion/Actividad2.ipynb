```python
import pandas as pd
import matplotlib.pyplot as plt


url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
titanic = pd.read_csv(url)
```

Analisis del contenido de la base de datos

```python
#primeras filas del dataset
titanic.head()
```

```python
#informacion geneal
titanic.info()
```

```python
#Datos que son nulos
titanic.isnull().sum()
```

# Analisis de variables categoricas

```python
varCateogricas = titanic.select_dtypes(include=["object","category"]).columns
varCateogricas.tolist()
```

```python
for var in varCateogricas:
    print(titanic[var].value_counts())

    plt.figure()
    titanic[var].value_counts().plot(kind="bar")
    plt.title(var)
    plt.show()
```

# Analisis de var5iables numericas


```python
varNumericas = titanic.select_dtypes(include=["int64","float64"]).columns
varNumericas.tolist()
```

```python
titanic[varNumericas].describe()
```

```python
titanic[varNumericas].hist(figsize=(15,10))
plt.show()
```

Comparacion y Visualizacion

```python
#Supervivencia vs Sexo
plt.figure()
titanic.groupby("Sex")["Survived"].mean().plot(kind="bar")
plt.show()
```

```python
#edad vs Supervicencia
plt.figure()
titanic.groupby("Age")["Survived"].mean().plot(kind="bar")
plt.show()
```

```python
#Supervivencia vs clase
```

```python
plt.figure()
titanic.groupby("Pclass")["Survived"].mean().plot(kind="bar")
plt.show()
```
