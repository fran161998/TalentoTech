Operaciones entre conjuntos

```python
#Union
conjunto1 = {1,2,3,4,5}
conjunto2 = {6,7,8,9,10}
conjuntoUnion = conjunto1.union(conjunto2)
print(conjuntoUnion)
```

```python
#Interseccion
conjunto3 = {1,2,3,4,5}
conjunto4 = {3,4,5,6,7}
conjuntoInterseccion = conjunto3.intersection(conjunto4)
print(conjuntoInterseccion)
```

```python
#Diferencia
conjunto5 = {1,2,3,4,5}
conjunto6 = {3,4,5,6,7,8,9}
conjuntoDiferencia5 = conjunto5.difference(conjunto6)
conjuntoDiferencia6 = conjunto6.difference(conjunto5)
print(conjuntoDiferencia5)
print(conjuntoDiferencia6)
```

```python
#Diferencia Simetrica
conjunto7 = {1,2,3,4,5}
conjunto8 = {3,4,5,6,7,8,9}
conjuntoDiferenciaSimetrica = conjunto7.symmetric_difference(conjunto8)
print(conjuntoDiferenciaSimetrica)
```

```python
#Complemento
conjunto9 = {1,2,3,4,5}
conjunto10 = {3,4,5,6,7,8,9}
conjuntoComplemento = conjunto9.difference(conjunto10)
print(conjuntoComplemento)
```

Diccionario

```python
futbolistasFCB = {"Portero":"TerStegen", "Defensa":"Cubarsi", "MedioCampista":"Pedri", "Delantero":"LamineYamal"}
print(futbolistasFCB)
print(futbolistasFCB["Portero"])
print(futbolistasFCB["MedioCampista"])
print(futbolistasFCB.keys())
print(futbolistasFCB.values())

for posicion, nombre in futbolistasFCB.items():
  print("Posicion: " + posicion +", Nombre: " + nombre)
```
