```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
x1 = [7, 1, 11, 11, 7, 11, 3, 1, 2, 21, 1, 11, 10]
x2 = [26, 29, 56, 31, 52, 55, 71, 31, 54, 47, 40, 66, 68]
x3 = [6, 15, 8, 8, 6, 9, 17, 22, 18, 4, 23, 9, 8]
x4 = [60, 52, 20, 47, 33, 22, 6, 44, 22, 26, 34, 12, 12]
y = [78.5, 74.3, 104.3, 87.6, 95.9, 109.2, 102.7, 72.5, 93.1, 115.9, 83.8, 113.3, 109.4]

# Create a pandas DataFrame
Data = pd.DataFrame({'x1': x1, 'x2': x2, 'x3': x3, 'x4': x4, 'y': y})

# Display the DataFrame
print(Data)
```

```python
y = Data['y']
X = Data[['x1', 'x2', 'x3', 'x4']]

X = sm.add_constant(X)

model = sm.OLS(y, X)
results = model.fit()

print(results.summary())
```

```python
predicted_values = results.predict(X)
residuals = results.resid


plt.figure(figsize=(8, 6))
plt.scatter(predicted_values, residuals, facecolors='red', edgecolors='black')
plt.hlines(0, predicted_values.min(), predicted_values.max(), colors='blue', linestyles='dashed', lw=1.5)
plt.vlines(predicted_values, 0, residuals, colors='red', lw=2, linestyle='-')

plt.title("Grafica")
plt.xlabel("Prediccion")
plt.ylabel("Residuo del Modelo")
plt.grid(True)
plt.show()
```
